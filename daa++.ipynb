```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os

# Load the image
img = Image.open("/content/drive/MyDrive/daa++/istockphoto-1208752997-1024x1024.jpg")

# Display the original image
plt.imshow(img)
plt.title("Original Image")
plt.show()

# Convert the image to a numpy array
img_array = np.array(img)
bit_rates = []
visual_errors = []
width, height = img.size
# Compress the image using JPEG with different quality ratios
for q in range(10, 101, 10):
    # Set the save path for the compressed image
    save_path = f"/content/drive/MyDrive/daa++/compressed_image_q{q}.jpg"
    
    # Compress the image with the specified quality ratio
    img_compressed = Image.fromarray(img_array)
    img_compressed.save(save_path, optimize=True, quality=q)
    
    # Load the compressed image and calculate its size
    img_compressed = Image.open(save_path)
    file_size = img_compressed.size[0] * img_compressed.size[1] * 3  # 3 channels (RGB)
    # Calculate bit rate and visual error
    compressed_size = os.path.getsize(save_path)
    bit_rate = compressed_size * 8 / (width * height)
    # Display the compressed image
    plt.imshow(img_compressed)
    plt.title(f"Compressed Image (q={q})")
    plt.show()
    
    # Calculate the visual error (MSE)
    mse = np.mean((np.array(img) - np.array(img_compressed)) ** 2)
    visual_error = mse / np.mean(np.array(img) ** 2) * 100
    
    # Print the file size and visual error
    print(f"Quality Ratio: {q}")
    print(f"File Size: {file_size} bits")
    print(f"Visual Error: {visual_error:.2f}")
    bit_rates.append(bit_rate)
    visual_errors.append(visual_error)
    
# Plot rate/distortion curve
plt.plot(bit_rates, visual_errors)
plt.xlabel("Bit rate (bpp)")
plt.ylabel("Visual error (0-10)")
plt.title("Rate/Distortion Curve")
plt.show()
```
